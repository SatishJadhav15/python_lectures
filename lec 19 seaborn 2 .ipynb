# Bar plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=sns.load_dataset('penguins')
```

```python
df1
```

```python
df1['island']
```

```python
sns.barplot(x=df1.island, y =df1.bill_length_mm)
plt.show()
```

```python
# or 
sns.barplot(x ='island', y= 'bill_length_mm', data= df1)
plt.show()
```

```python
sns.barplot(x= 'island', y ='bill_length_mm',data= df1, hue = 'sex')
plt.show()
```

# Order

```python
order1 = ['Dream', 'Torgersen','Biscoe']
sns.barplot(x='island', y='bill_length_mm', hue='sex', data=df1, order= order1)
plt.show()
```

# hue_order

```python
order1 = ['Dream', 'Torgersen','Biscoe']
sns.barplot(x='island', y='bill_length_mm', hue='sex', data=df1, 
            order= order1, hue_order= ['Female','Male' ], palette= ['orange','green'])
plt.show()
```

```python
# single colour
sns.barplot(x='island', y='bill_length_mm',data =df1 ,
        color='pink')
plt.show()
```

```python
sns.barplot(x= 'island', y= 'bill_length_mm',data= df1, hue='sex', palette= 'Accent')
plt.show()
```

```python
sns.barplot(x= 'island', y= 'bill_length_mm',data= df1, hue='sex', palette= 'prism')
plt.show()
```

# Saturation

```python
# sasturatuion parameeter used to dull, fade the colour
sns.barplot(x= 'island', y= 'bill_length_mm',data= df1, hue='sex', palette= 'prism', saturation =0.4)
plt.show()
```

# Histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1
```

```python
sns.displot(df1['flipper_length_mm'])
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'], bins =[170,180,190,200,210,220,230,240])
plt.show()
```

```python
sns.displot(df1['flipper_length_mm'],bins =[170,180,190,200,210,220,230,240], color ='g', kde= True)
#here bins is provided to distinguish the histogram in between these provided values
```

 # kde

```python
# kde = kernal density function
sns.displot(df1['flipper_length_mm'],  bins =[170,180,190,200,210,220,230,240], color ='g', kde=True)
plt.show()
```
