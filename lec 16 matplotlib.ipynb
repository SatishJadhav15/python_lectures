# Matplotlib library

used to convert numerical data into visualisation like bar, scatter, linear graph

```python
import matplotlib.pyplot as plt
```

# Linear graph

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)   # plot for linear line
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
c ='g'    # to give colour to graph

plt.plot(x,y,c)
plt.show()
```

```python
x = [2,1,5,9,10]
y = [10,52,89,63,82]

plt.plot(x,y)
plt.show()
```

```python
x = [2,1,5,9,10]
y = [10,52,89,63,82]

plt.plot(x,y, marker = 'o')   # marker is used get dot at corner points
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2,1,5,9,10]
y = [10,52,89,63,82]

plt.plot(x,y)
plt.plot(y, marker ='<')           #graph of y axis condsidering continuous numbers of x like 0 1 2 3 4  etc   means graph of y, indepedent of x
plt.plot(x, marker ='o')   
plt.show()
```

# Bar graph

```python
x = [1,2,3,4]
y= [5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y= [5,6,7,8]
c = ['g','b','y','r']
plt.bar(x,y, color = c)
plt.show()
```

```python
import numpy as np
```

```python
x =np.array(['A','B','C','D'])
y =np.array([10,50,80,60])

plt.barh(x,y, color = 'red')   # h for horizontal bar graph
plt.show()
```

```python
x =np.array(['A','B','C','D'])
y =np.array([10,50,80,60])

plt.bar(x,y, color = 'green', width = 0.3)   # h for horizontal bar graph
plt.show()
```

```python
x =np.array(['A','B','C','D'])
y =np.array([10,50,80,60])

plt.barh(x,y, color = 'lightblue',height = 0.1)   # h for horizontal bar graph
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [6,5,9,8,5]
plt.title('Scatter plot')
plt.ylabel("number")
plt.xlabel("month")

plt.scatter(x,y,color = 'red')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,5,9,8,5]
plt.title('Scatter plot')
plt.ylabel("number")
plt.xlabel("month")

c =['green', 'magenta', 'yellow', 'red', 'blue']

plt.scatter(x,y,color = c)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [6,5,9,8,5]
plt.title('Scatter plot')
plt.ylabel("number")
plt.xlabel("month")

c =['green', 'magenta', 'yellow', 'red', 'blue']

plt.scatter(x,y,color = c, s = 200)    #  s is for to increase the size of ball
plt.show()
```
