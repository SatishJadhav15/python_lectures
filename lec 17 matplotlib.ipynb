```python
# importing libraries
# converting image into black and white

import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

fname = r'india.png'

# opening image into pil

image = Image.open(fname).convert('L')

#mapping image into gray scale
plt.imshow(image, cmap = 'gray')
plt.show()
```

```python
# coverting into prescribe color
import matplotlib.pyplot as plt 
import pandas as pd
from PIL import Image

fname = r'mountain.jpeg'

image = Image.open(fname).convert('L')
plt.imshow(image, cmap = 'RdBu_r')      # color names are available just do wrong and will all the colors
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image

fname = r'modi.jpg'
image = Image.open(fname).convert('L')
plt.title("Narendra Modi")
plt.xlabel('Length')
plt.ylabel('Height')
plt.grid()
plt.imshow(image, cmap = 'gray')
plt.show()
```
