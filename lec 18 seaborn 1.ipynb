```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
var =[1,2,3,4,5,6]
var1= [8,7,9,4,8,7]
```

# matplotlib

```python
plt.plot(var,var1)
plt.show()
```

# seaborn

```python
sns.lineplot(x = var, y= var1)
plt.show()
```

```python
var =[1,2,3,4,5,6]
var1= [8,7,9,4,8,7]

import pandas as pd

df =pd.DataFrame({'var':var,'var1':var1})
sns.lineplot(x="var", y="var1" , data= df)
plt.show()
```

```python
df1 = sns.load_dataset("penguins")
```

```python
df1
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=df1)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=df1, hue ='sex')  # hue is for categorical selection
plt.show()                                                            # data means form which data set we are importing the values
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=df1, hue ='sex', palette = 'rainbow') 
plt.show()                                               # pallete means giving the color
```

```python
sns.lineplot(x='bill_length_mm', y='bill_depth_mm', data=df1, hue ='sex',
             palette = 'rainbow',  style= 'sex', markers = ['o','>']) 
plt.show()        
```
